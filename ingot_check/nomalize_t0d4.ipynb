```c++
TFile *f = TFile::Open("/data/disk1/ribll2026_www_data/normalize/t0d4_t1_0060_0060.root");
TFile *opt = new TFile("/home/ribll2026/ribll2026_www/github_code/jupyter/ingot_check/root_file/normalize_t0d4.root","recreate");

TGraph *gf[32];
TH1D *resf[32];
TGraph *gb[32];
TH1D *resb[32];

for (int i = 0; i < 32; i++) {
    // 创建 canvas，只在循环内使用，不删除
    TCanvas *c1 = new TCanvas(Form("c1_%d",i), Form("front_back_%d",i), 1000, 500);
    TCanvas *c2 = new TCanvas(Form("c2_%d",i), Form("front_back_%d",i), 1000, 500);
    c1->Divide(2,1);
    c2->Divide(2,1);

    // 直接从文件读取（不需要先 new）
    gf[i] = (TGraph*)f->Get(Form("gf%d",i));
    resf[i] = (TH1D*)f->Get(Form("resf%d",i));
    
    // 检查是否读取成功
    if (!gf[i] || !resf[i]) {
        cout << "Warning: Cannot find gf" << i << " or resf" << i << endl;
        continue;
    }
    
    c1->cd(1);
    gf[i]->SetTitle(Form("front strip %d",i));
    gf[i]->Draw("AP*");
    c1->cd(2);
    resf[i]->Draw();

    gb[i] = (TGraph*)f->Get(Form("gb%d",i));
    resb[i] = (TH1D*)f->Get(Form("resb%d",i));
    
    if (!gb[i] || !resb[i]) {
        cout << "Warning: Cannot find gb" << i << " or resb" << i << endl;
        continue;
    }
    
    c2->cd(1);
    gb[i]->SetTitle(Form("back strip %d",i));
    gb[i]->Draw("AP*");
    c2->cd(2);
    resb[i]->Draw();

    // 写入 canvas 到文件
    opt->cd();  // 切换到输出文件
    c1->Write(Form("front%02d",i));
    c2->Write(Form("back%02d",i));
    
    // 删除 canvas（注意：不要删除从 f 读取的对象）
    delete c1;
    delete c2;
}

opt->Close();
// f->Close();  // 记得关闭输入文件
```
